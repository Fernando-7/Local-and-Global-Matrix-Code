```python
import numpy as np #Código de matriz global

def Global(arquivo,arquivo2):  
    
    arq = open(arquivo,'r')   # abrindo arquivos das sequencias do pares de moedas
    arq2 = open(arquivo2,'r')
    
    novo = open('matrizSequencia.txt','w+') # crinado arquivo onde ficara a matriz
    
    cadeia = arq.read()
    cadeia2 = arq2.read()
    
    mat_global = ['0'] * int(len(cadeia)) # criando matriz de caracteres para inserir no arquivo
    for i in range(int(len(cadeia))):
        mat_global[i] = ['0'] * int(len(cadeia2))
    
    matriz = np.empty((len(cadeia), len(cadeia2)))                
    
    for i in range(0,len(cadeia)):  #apenas para sabermos onde cada posição de coluna representa a primeira string
        matriz[i][0] = 0
        mat_global[i][0] = cadeia[i]
        
    for i in range(0,len(cadeia2)): #apenas para sabermos onde cada posição de linha representa a primeira string    
        matriz[0][i] = 0
        mat_global[0][i] = cadeia2[i]
        
    for i in range(1,len(cadeia)):
        for j in range(1,len(cadeia2)):
            
            match = int(cadeia[i] == cadeia2[j])
            
            matriz[i][j] = max((matriz[i-1][j] - 1), (matriz[i-1][j-1] + match), (matriz[i][j-1] - 1)) #seleciona o valor máximo de acordo com a regra do sequenciamento
            mat_global[i][j] = str(int(matriz[i][j]))
            novo.write(mat_global[i][j]+" ") # escreve no arquivo
    novo.write("\n")
    
    novo.close()
    arq.close()
    arq2.close()
    
    print(mat_global) 

#Global();
```
