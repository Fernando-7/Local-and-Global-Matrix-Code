```python
import matplotlib.pyplot as plt
import csv
ficheiro = open('EURFCH1H.csv')
vet=[]
reader = csv.reader(ficheiro, delimiter=',')
for linha in reader:
    if ((reader.line_num) < 160):
        vet.append(linha[2])
plt.plot(vet)
plt.show()
    
#plt.plot([1,2,3])
#plt.title("Teste")
#plt.show()
```
